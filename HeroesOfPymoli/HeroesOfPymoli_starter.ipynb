## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Bins and labels 
bins = [0,9, 14, 19, 24, 29, 34, 39, 900]
bin_names = ['<10', '10-14', '15-19', '20-24', '25-29','30-34','35-39','40+']

# bin the data based on age and calculate the percentages and write to a new dataframe
bin_age=pd.DataFrame((((pd.cut(purchase_data['Age'], bins, labels=bin_names)).value_counts())/total_p*100))

# Renaming the new dataframe column.
bin_age.rename(columns={'Age': 'Percentage of Players'},inplace = True)

# Total count based age
bin_age['Total Count']=pd.cut(purchase_data['Age'], bins, labels=bin_names).value_counts()

#Puting bin_names in order
bin_age=bin_age.reindex(bin_names)

# Formating and outpout
bin_age.style.format({'Total Count': '{:.0f}',
                    'Percentage of Players': "{:.0f}%", })\
    .set_properties(**{'background-color': 'White',
                                    'color': 'black',
                                    'text-align': 'center',
                                    'border-color': 'black',
                                    'border-width': '2px',
                                    'border-style': 'solid'})\
                                    .bar(subset=['Total Count'], color='lightblue')\
                                    .bar(subset=['Percentage of Players'], color='lightgreen')\
                                    .highlight_min(subset=None, color='salmon', axis=0)\
                .set_caption('Age Demographics')
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
 # Id counts from purchase data for each SN
top_s = purchase_data['Item ID'].groupby(purchase_data['SN']).count()

# Created column in the Dataframe
top_s = pd.DataFrame(data = top_s)
top_s.columns = ['Purchase Count']

# The average purchase price per screen name
top_s['Average Purchase Price'] = purchase_data['Price'].groupby(purchase_data['SN']).mean()

# The total purchase price per screen name
top_s['Total Purchase Value'] = purchase_data['Price'].groupby(purchase_data['SN']).sum()

# Sort by descending order
top_s.sort_values(by=['Total Purchase Value'], ascending = False, inplace = True)

# Formating and outpout
top_s.head(10).style.format({'Purchase Count': '{:.0f}',
                                'Average Purchase Price': '${:.2f}',
                                'Total Purchase Value': '${:.2f}'})\
                        .set_properties(**{'background-color': 'white',      
                                        'color': 'black',
                                        'text-align': 'center',
                                        'border-color': 'black',
                                        'border-width': '2px',
                                        'border-style': 'solid',})\
                        .background_gradient()\
                        .highlight_min(subset=None, color='salmon', axis=0)\
                        .set_caption('Top Spenders')
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
p_items = purchase_data.filter(['Item ID', 'Item Name','Price'], axis=1)
item_price = p_items.groupby(['Item ID','Item Name'])['Price'].mean()
purchase_count = p_items.groupby('Item ID')['Item Name'].value_counts()
purchase_value = item_price * purchase_count

popular = purchase_count.to_frame('Purchase Count')
popular['Item Price'] = item_price
popular['Total Purchase Value'] = purchase_value
popular1 = popular.sort_values(by="Purchase Count", ascending=False)

popular1.head()
                        
```

* Sort the above table by total purchase value in descending order

```python
# Sorting the most profitable in descending order
prof_items = popular.sort_values(by='Total Purchase Value', ascending=False)

# Formating and outpout
prof_items.head(10).style.format({'Purchase Count': '{:.0f}',
                                'Item Price': '${:.2f}',
                                'Total Purchase Value': '${:.2f}'})\
                        .set_properties(**{'background-color': 'white',      
                                        'color': 'black',
                                        'text-align': 'center',
                                        'border-color': 'black',
                                        'border-width': '2px',
                                        'border-style': 'solid',})\
                        .background_gradient()\
                        .highlight_min(subset=None, color='salmon', axis=0)\
                        .set_caption('Most Profitable Items')
```
